2. Compute Friday-to-Friday returns using the log difference of closing prices and produce a histogram. Experiment with the number of bins.

```python
snp_data_dt = snp_data.copy()
```

```python
snp_data_dt['Date'] = x_values
```

```python
# Checking if it's a Friday:
snp_data_dt['Date'][8].weekday() == 4
```

```python
snp_data_dt['Date'][12].weekday() == 4
```

```python
# Option 1:
snp_fridays = []
for i in range(snp_data_dt.shape[0]):
    if snp_data_dt['Date'][i].weekday() == 4:
        snp_fridays.append(snp_data_dt.loc[i, 'Adj Close'])
```

```python
# Calculating log difference:
snp_log_returns = (np.log(snp_fridays[1:]) - np.log(snp_fridays[:-1]))
plt.hist(snp_log_returns, bins=50)
plt.show()
```

```python
# Option 2:
snp_data_dt['Is_Friday'] = False
for i in range(snp_data_dt.shape[0]):
    if snp_data_dt['Date'][i].weekday() == 4:
        snp_data_dt.loc[i, 'Is_Friday'] = True
```

```python
snp_data_dt
```

```python
snp_fridays_1 = snp_data_dt['Adj Close'][snp_data_dt['Is_Friday']]
```

```python
# Calculating log difference:
snp_log_returns = (np.log(snp_fridays_1) - np.log(snp_fridays_1.shift(1))).dropna()
plt.hist(snp_log_returns, bins=50)
plt.show()
```

3. Compute the percentage of weekly returns and produce a pie chart containing the percentage of weekly returns in each of:

```python
pct_change = snp_fridays_1.pct_change().dropna()
```

```python
pie_cats = np.zeros((4))
pie_cats[0] = np.sum(pct_change<=-0.02)
pie_cats[1] = np.sum((pct_change>-0.02)&(pct_change<=0))
pie_cats[2] = np.sum((pct_change>0)&(pct_change<=0.02))
pie_cats[3] = np.sum(pct_change>0.02)
```

```python
# Check point, should be true:
int(np.sum(pie_cats)) == len(snp_fridays_1) - 1
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 15))
ax.pie(pie_cats, labels=['<-2%','>-2% and <0','>0 and <2%','>2%'], explode = [0,0,0.2,0], autopct = '%2.0f', shadow = True)
plt.show()
```
