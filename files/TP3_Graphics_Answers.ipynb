6. Compute EWMA variance for both the S&P500 and FTSE and plot against dates. An EWMA variance has $\sigma^2_{t'} =(1âˆ’Î»)r^2_{t-1} + \lambda \sigma^2_{t-1}$ where $r^2_0 =Ïƒ^2_0$ is the full sample variance and $\lambda=0.97$.

```python
returns = np.stack([snp_log_returns_10[nonnan_indices], ftse_log_returns[nonnan_indices]])
ewma = np.zeros((returns.shape))
ewma[:,0] = np.var(returns, axis=1)
l = 0.97
for i in range(1,returns.shape[1]):
    ewma[:,i] = (1-l)*returns[:,i-1] + l*ewma[:,i-1]
x_values_10 = ftse_log_returns[nonnan_indices].index
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 7))
plt.plot(x_values_10, ewma[0], '--', markersize=4)
plt.plot(x_values_10, ewma[1], '-.', markersize=4)
ax.legend(['S&P 500 EWMA', 'FTSE 100 EWMA'])
years = mpl.dates.YearLocator()
ax.xaxis.set_major_locator(years)
fmt = mpl.dates.DateFormatter('%b %Y')
ax.xaxis.set_major_formatter(fmt)
fig.autofmt_xdate()
plt.tight_layout()
plt.draw()
```
