1. For each of the following distributions, simulate 1000 pseudo-random numbers:

```python
normal_st = np.random.standard_normal((1000))
plt.hist(normal_st, bins=30, alpha=0.7)
normal_1 = np.random.normal(3, 3, (1000))
plt.hist(normal_1, bins=30, alpha=0.7)
plt.show()
```

```python
unif_1 = np.random.uniform(0, 1, (1000))
unif_2 = np.random.uniform(-1, 1, (1000))
plt.hist(unif_1, bins=30, alpha=0.7)
plt.hist(unif_2, bins=30, alpha=0.7)
plt.show()
```

```python
gamma = np.random.gamma(1, 2, (1000))
plt.hist(gamma, bins=30, alpha=0.7, density=True)
plt.plot(np.linspace(0, 12, 100), sps.gamma.pdf(np.linspace(0, 12, 100), 1, scale=2), c='r')
plt.show()
```
