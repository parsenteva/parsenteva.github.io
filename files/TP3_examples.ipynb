```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

## Line plots:

```python
y = np.random.randn(100)
```

```python
plt.plot(y)
```

```python
x = np.linspace(0,1,100)
```

```python
x
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'g--')
```

```python
plt.plot(x,y,alpha = 0.5, color = '#FF7F00', label = 'Line Label', linestyle = '-.',linewidth = 3, marker = 'o', markeredgecolor = '#000000', markeredgewidth = 2, markerfacecolor = '#FF7F00',markersize=30)
```

## Scatter plots:

```python
z = np.random.randn(2, 100)
```

```python
plt.scatter(z[0], z[1])
```

## Subplots:

```python
fig, axs = plt.subplots(1, 2, figsize=(10,5))
axs[0].plot(x, y, 'g--')
axs[0].plot(x, z[0], 'b--')
axs[0].legend(['green dashy line'])
axs[0].set_title('Plot 1')
axs[1].scatter(z[0], z[1], color='r')
axs[1].legend(['red points'])
axs[1].set_title('Plot 2')
fig.suptitle('Plots')
plt.show()
```

## Bar plots:

```python
y = np.random.rand(5)
x = np.arange(5)
plt.bar(x, y, width = 0.5, color = '#FF7F00', edgecolor = '#000000', linewidth = 5)
```
